Read from update DataFrame

```python
df=pd.read_csv('yogesh.csv')
df
```

clean up data
Display all rows of NAN

```python
df=df.dropna(how='all')
df
```

```python
df.info()
```

eliminate rows

```python
df=df.loc[df['Quantity Ordered']!='Quantity Ordered']
df
df.to_csv('yogesh.csv')
```

```python
df.info()
```

```python
df['Order Date']=pd.to_datetime(df['Order Date'])
df['Quantity Ordered']=df['Quantity Ordered'].astype(int)
df['Price Each']=df['Price Each'].astype(float)
```

```python
df.info()
```

```python
df['Amount']=df['Quantity Ordered']*df['Price Each']
df['Month']=df['Order Date'].dt.month
df
```

What was the best month for sales? How much was earned that month?

```python
ndf=df.groupby('Month').agg(top_sales_in_month=('Amount','sum'))
ndf=ndf.sort_values(by='top_sales_in_month',ascending=False)
ndf=pd.DataFrame(ndf)
ndf=ndf.reset_index()
ndf
```

plot the data

```python

def ctreatebar(x,y,xlabe='',ylabe='',labe='',size=(10,5),xticks='',titl='',siz=(10,5)):
    plt.figure(figsize=siz)
    plt.bar(x,y,label='labe')
    plt.xlabel(xlabe)
    plt.xticks(xticks)
    plt.title(titl)
    plt.ylabel(ylabe)
    plt.legend()
    plt.show()
```

```python
x=ndf['Month']
y=ndf['top_sales_in_month']
ctreatebar(x=x,y=y,xlabe='Month',ylabe='Sales',titl='top sales in month',siz=(10,5),xticks=np.arange(1,13,1),labe='top sales')
```

What city sold the most product?

```python
df['City']=df['Purchase Address'].str.split(',',expand=True)[1]
df
```

```python
sdf=df.groupby('City').agg(Top_product_in_city=('Quantity Ordered','sum'))
sdf=sdf.sort_values(by='Top_product_in_city',ascending=False)
sdf=pd.DataFrame(sdf)
sdf=sdf.reset_index()
sdf
```

```python
a=sdf['City']
b=sdf['Top_product_in_city']
ctreatebar(x=a,y=b,xlabe='City',ylabe='Sales in USD',titl='top product in city',xticks=np.arange(0,9,1),siz=(15,5))
```

What time should we display advertisements to maximize the likelihood of customer’s buying product?

```python
df['Hour']=df['Order Date'].dt.hour
df
```

```python
hdf=df.groupby('Hour').agg(time_people_bought=('Amount','sum'))
hdf=hdf.sort_values(by='time_people_bought',ascending=False)
hdf=pd.DataFrame(hdf)
hdf=hdf.reset_index()
hdf
```

```python
hdf.columns
```

plot the data

```python
sns.scatterplot(data=hdf,x='Hour',y='time_people_bought')
sns.lineplot(data=hdf,x='Hour',y='time_people_bought')
plt.grid()
plt.xticks(ticks=np.arange(0,24,1))
plt.show()
```

What Products are most often sold together 

```python
df.to_csv('yogesh1.csv')
ydf=pd.read_csv('yogesh1.csv')
ydf=ydf.drop(columns='Unnamed: 0.1')
ydf=ydf.loc[ydf['Order ID'].duplicated(keep=False)]
ydf
```

Group Different Product with same order id

```python
S=ydf.groupby('Order ID').agg( result=('Product','sum'))
S
```

```python
ydf['Group']=S['result']
ydf
```

```python
ydf['group']=ydf.groupby('Order ID')['Product'].transform(lambda x:','.join(x))
ydf
```

```python
from itertools import combinations
from collections import Counter
count=Counter()
row_list=[]
for row in ydf['group']:
    row_list=row.split(',')
    count.update(Counter(combinations(row_list,3)))

count.most_common(10)
```
